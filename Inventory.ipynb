```
import numpy as np
import matplotlib.pyplot as plt  
```

```
m = 11 #maximum capacity
n = 5 #review length

begining_inventory= 3
demand=0
ending_inventory=0
order_quantity=8
shortage_quantity=0
days_until_order_arrives=2
days=0;
day_no=[]
ending=[]
for cycle in range (10):
  print("Cycle no:", cycle+1)
  for day in range (1,n+1):
    days=days+1;
    day_no.append(days)
    print("Day no; ",day)
    days_until_order_arrives-=1
    if days_until_order_arrives==-1:
      begining_inventory=begining_inventory+order_quantity
      begining_inventory=begining_inventory-shortage_quantity
      shortage_quantity=0


    daily_demand= np.random.choice(a=[0,1,2,3,4],p=[.10,.25,.35,.21,.09])
    total_demand = daily_demand + shortage_quantity
    if total_demand> begining_inventory:
      shortage_quantity=total_demand-begining_inventory
      ending_inventory=0
    else:
      ending_inventory=begining_inventory-total_demand
    ending.append(ending_inventory)
    print("Begining Inventory:",begining_inventory)
    print("Daily Demand:",daily_demand)
    print("Ending Inventory:",ending_inventory)
    print("Shortage Quantity:",shortage_quantity)
    print()
    
    begining_inventory=ending_inventory
    if day==n:
      order_quantity=m-begining_inventory
      days_until_order_arrives= np.random.choice(a=[1,2,3],p=[.60,.30,.10])


  

plt.bar(day_no,ending,width = 0.8,color = ['red', 'green']) 
plt.xlabel('Days Number') 
plt.ylabel('Ending Inventory') 
plt.title('Inventory VS Day Graph') 
plt.show() 

```
