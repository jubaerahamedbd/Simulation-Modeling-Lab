```python
import random
import math
import matplotlib.pyplot as plt
n = [10]
x2=[]
y2=[]
x3=[]
far=0
m=0
r=0
xL=0
fa=[]
ra=[]
ma=[]
xR=100
trials =5
farArrayX=[]
farArrayY=[]
midArrayX=[]
midArrayY=[]
randArrayX=[]
randArrayY=[]
targetIndex=0
targetList =[500,1000,1500,2000,2500]
rand=1
farth=1
middle=1
middleArray=[]
nearLow=0
nearHigh=5
mL=0
mR=100
minimumList=[]
rL=0
rR=100
randArray=[]
finalrandArray=[]
arrayvalue=0
for j in range (100):
  for i in range(trials):
      x= random.uniform(xL,xR)
      mx= random.uniform(mL,mR)
      rx= random.uniform(rL,rR)
      y= random.uniform(0,5)
      minimumList.append(mx)
      x2.append(x)
      x3.append(rx)
      y2.append(y)   

  x2.sort()
  far=max(x2)
  far_index=[x2.index(far)]
  xL=far
  xR=far+100
  farArrayX.append(x2[x2.index(far)])
  farArrayY.append(y2[x2.index(far)])
  #print(farArrayX)
  minimumList.sort()
  middleArray = minimumList[nearLow:nearHigh]
  
  
  m1=min(middleArray)
  #print(m1)
  minimumIndex=[minimumList.index(m1)]
  mL=m1
  mR=m1+100
  #print(m1)
  m1_index=[minimumList.index(m1)]
  midArrayX.append(minimumList[minimumList.index(m1)])
  #print(midArrayX)
  #midArrayY.append(y2[x3.index(m1)])

  rendItem=random.randint(0, 4)
  randArray= x3[nearLow:nearHigh]
  #print(randArray)
  #print(x3)
  #print(randArray[rendItem])
  arrayvalue=randArray[rendItem]
  #break
  #print(rendItem)
  nearLow=nearLow+5
  nearHigh=nearHigh+5
  #print(arrayvalue)
  finalrandArray.append(arrayvalue)
  #print(finalrandArray)
  
  #break
  rL=arrayvalue
  rR=arrayvalue+100
 # print(randArrayX)
  

  if (rand ==1):
    if (max(finalrandArray) >= targetList[targetIndex]): 
      print("Random   reached", targetList[targetIndex] ,"m at",j, "times Broadcting")
      ra.append(j) 
      rand=0
      #break
  if (middle == 1):
    if (max(midArrayX) >= targetList[targetIndex]): #and (max(midArrayX) <= i+50):
      print("Nearpoint reached", targetList[targetIndex] ,"m at",j, "times Broadcting")
      ma.append(j)  
      middle=0   
      #break
  if (farth ==1 ):
    if (max(farArrayX) >= targetList[targetIndex]):#and (max(farArrayX) <= i+50):
      print("Farpoint reached", targetList[targetIndex] ,"m at",j, "times Broadcting")
      fa.append(j)
      farth=0
      #break
  if (rand==0 and middle==0 and farth ==0):
    rand=1
    middle=1
    farth=1
    targetIndex=targetIndex+1;
    if (targetIndex == len(targetList)):
      break

print('fa',fa)
print('ra',ra)
print('ma',ma)

plt.plot(targetList,fa,color="green") 
plt.plot(targetList,ra,color="red")
plt.plot(targetList,ma,color="Black")

plt.legend(["Near Points","Random Points","Far Points"])      # plot the value in a graph
plt.xlabel('Number of Trials') 
plt.ylabel('Number of Broadcasting') 
#plt.title('After 20 time Broadcasting:') 
#values=['1','2','3','4','5']
#plt.xticks(x,values)
plt.show() 
```
